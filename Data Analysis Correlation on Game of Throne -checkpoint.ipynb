#### Converting the Categorical Variables to Numerical variables

```python
df_convert = df
for categ in df_convert.columns:
    if df_convert[categ].dtype == 'object':
        df_convert[categ] = df_convert[categ].astype('category')
        df_convert[categ] = df_convert[categ].cat.codes
    
df_convert
```
